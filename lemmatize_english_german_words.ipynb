```python
import spacy
nlp_en = spacy.load("en_core_web_sm")
```

```python
path = 'words/german_top_1000_translated.txt'
# strip removes leading and trailing whitespaces
with open(path, 'r') as file:
    words = file.readlines()
    words = [word.strip() for word in words]
```

```python
lemmatized_words = []
for word in words:
    item = nlp_en(word)
    if len(item) > 1 or len(item) == 0:
        lemmatized_word = 'oov'
    else:
        first_word = item[0]
        lemmatized_word = first_word.lemma_.lower()
    lemmatized_words.append(lemmatized_word)
```

```python
#count number of oov
oov = 0
for word in lemmatized_words:
    if word == 'oov':
        oov += 1
print(oov)
```

```python
# save
save_dest = 'words/german_top_1000_translated_lemmatized.txt'
with open(save_dest, 'w') as file:
    for word in lemmatized_words:
        file.write(word + '\n')
```
