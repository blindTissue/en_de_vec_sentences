```python
from gensim.models import Word2Vec
import os
from tqdm import tqdm

english_model = Word2Vec.load(os.path.join('data', 'english_model_lemmatized'))
german_model = Word2Vec.load(os.path.join('data', 'german_model_lemmatized'))
```

```python
#save top 2000 words as text files
english_top_2000= english_model.wv.index_to_key[:2000]
german_top_2000 = german_model.wv.index_to_key[:2000]

with open(os.path.join('data', 'english_top_2000.txt'), 'w') as f:
    for word in english_top_2000:
        f.write(word + '\n')
        
with open(os.path.join('data', 'german_top_2000.txt'), 'w') as f:
    for word in german_top_2000:
        f.write(word + '\n')
```

```python
from deep_translator import GoogleTranslator
to_translate = 'kommission'
translated = GoogleTranslator(source='de', target='en').translate(to_translate)
```

```python
translated_de = []
for word in tqdm(german_top_2000):
    translated = GoogleTranslator(source='de', target='en').translate(word)
    translated_de.append(translated)
```

```python
for i in translated_de:
    print(i)
```

```python
print(len(translated_de))
print(len(set(translated_de)))
```

```python
for i in range(len(translated_de)):
    if translated_de[i] == None:
        translated_de[i] = 'none'
        
```

```python
import spacy
nlp_en = spacy.load('en_core_web_sm')
def lemmatize_en(text):
    return [tok.lemma_.lower() for tok in nlp_en(text) if not tok.is_punct]
```

```python
new_lemmatized_out = []
for i in range(len(translated_de)):
    print(i)
    new_lemmatized_out.append(lemmatize_en(translated_de[i]))
```

```python
save_items = []
for i in range(len(new_lemmatized_out)):
    if len(new_lemmatized_out[i]) > 1 or len(new_lemmatized_out[i]) == 0:
        save_items.append('multword')
    else:
        save_items.append(new_lemmatized_out[i][0])
        
```

```python
save_items
```

```python
#save words
with open(os.path.join('data', 'german_top_2000_translated_lemmatized.txt'), 'w') as f:
    for word in save_items:
        f.write(word + '\n')
```

```python
#create train set with same word within top 1000 words

english_top_1000 = english_model.wv.index_to_key[:1000]
german_top_1000 = save_items[:1000]
```

```python
train_set = []
for i in range(len(english_top_1000)):
    for j in range(len(german_top_1000)):
        if english_top_1000[i] == german_top_1000[j]:
            train_set.append((english_top_1000[i], german_top_1000[j], i, j))
            break
```

```python
print(len(train_set))
```

```python
train_set[:10]
```

```python
tot_set = []
for i in range(len(english_top_2000)):
    for j in range(len(save_items)):
        if english_top_2000[i] == save_items[j]:
            german_word = german_top_2000[j]
            tot_set.append((english_top_2000[i], german_word, i, j))
            break
```

```python
print(len(tot_set))
```

```python
#shuffle tot_set
import random
random.shuffle(tot_set)
```

```python
train_test_split = .8
train_set = tot_set[:int(len(tot_set) * train_test_split)]
test_set = tot_set[int(len(tot_set) * train_test_split):]
```

```python
# save
with open(os.path.join('train_test_data', 'train_set.txt'), 'w') as f:
    for item in train_set:
        f.write(item[0] + ' ' + item[1] + ' ' + str(item[2]) + ' ' + str(item[3]) + '\n')
        
with open(os.path.join('train_test_data', 'test_set.txt'), 'w') as f:
    for item in test_set:
        f.write(item[0] + ' ' + item[1] + ' ' + str(item[2]) + ' ' + str(item[3]) + '\n')
```

```python
#save tot_set
with open(os.path.join('train_test_data', 'tot_set.txt'), 'w') as f:
    for item in tot_set:
        f.write(item[0] + ' ' + item[1] + ' ' + str(item[2]) + ' ' + str(item[3]) + '\n')
```

```python
# order by the difference of the index
import math

tot_set_sorted = sorted(tot_set, key=lambda x: abs(x[2] - x[3]))
```
