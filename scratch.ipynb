```python
import nltk
from nltk.tokenize import word_tokenize

words = open('train_test_data/tot_set.txt').read().splitlines()
```

```python
words
```

```python
words = [word.split()[0] for word in words]
```

```python
words
```

```python
words = [word_tokenize(word) for word in words]
```

```python
words
```

```python
for i in words:
    print(nltk.pos_tag(i))
```

```python
# create counter
counts = {}
for i in words:
    item = nltk.pos_tag(i)[0][1]
    if item in counts:
        counts[item] += 1
    else:
        counts[item] = 1
    
    
```

```python
counts
```

```python
#order counts
counts = {k: v for k, v in sorted(counts.items(), key=lambda item: item[1], reverse=True)}
```
